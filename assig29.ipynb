```python
#16. Which video has the highest percentage of likes? like/(like+dislike)  * 100 
likep = []
for i in range(0, len(f)):
    likep.append((f['likes'][i]/(f['likes'][i] + f['dislikes'][i])) * 100)

j = likep.index(max(likep))
f['title'][j]
```

```python
#17. Which video has the highest percentage of dislikes?
dislikep = []
for i in range(0, len(f)):
    dislikep.append((f['dislikes'][i]/(f['likes'][i] + f['dislikes'][i])) * 100)

j = dislikep.index(max(dislikep))
f['title'][j]
```
